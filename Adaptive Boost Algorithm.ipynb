### Mushroom Hunting: Edible or Poisonous?

Data Source: https://archive.ics.uci.edu/ml/datasets/Mushroom


This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. This latter class was combined with the poisonous one. The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like ``leaflets three, let it be'' for Poisonous Oak and Ivy.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Mushrooms.csv")
```

```python
df.head()
```

```python
#  analyzing distribution
```

```python
sns.countplot(x='class',data=df)
```

```python
df.describe().transpose()
```

```python
feat_uni=df.describe().transpose().reset_index().sort_values('unique')
```

```python
plt.figure(figsize=(10,5),dpi=200)
sns.barplot(data=feat_uni,x='index',y='unique')
plt.xticks(rotation=90);
```

```python
X=df.drop('class',axis=1)
```

```python
X
```

```python
# Since In decision tree we  take dummies of these categorical features.so get dummies
```

```python
X=pd.get_dummies(X,drop_first=True,dtype=int)
```

```python
y=df['class']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model= AdaBoostClassifier(n_estimators=1)  # will split on basis of single best feature
```

```python
model.fit(X_train,y_train)
```

```python
y_pred= model.predict(X_test).reshape(-1,1)
```

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay,accuracy_score
```

```python
print(classification_report(y_test,y_pred))
```

```python
#  importance of features
model.feature_importances_
```

```python
model.feature_importances_.argmax()
```

```python
X.columns[22]
```

```python
#  now analyzing with more stumps
error_rates=[]

for n in range(1,96):
    model=AdaBoostClassifier(n_estimators=n)
    model.fit(X_train,y_train)
    
    y_pred=model.predict(X_test)
    error=1- accuracy_score(y_test,y_pred)
    error_rates.append(error)
```

```python
plt.plot(range(1,96),error_rates)
```

```python
model
```

```python
features=pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Importance'])
```
